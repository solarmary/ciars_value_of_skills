```python
def remove_outliers(df):
    mean_salary = df["salary_from_rub"].mean()
    std_salary = df["salary_from_rub"].std()
    lower_threshold = mean_salary - 3 * std_salary
    upper_threshold = mean_salary + 3 * std_salary
    return df[(df["salary_from_rub"] >= lower_threshold) & (df["salary_from_rub"] <= upper_threshold)]

df_filtered = df_demo.groupby("region_name").apply(remove_outliers).reset_index(drop=True)
df_filtered.shape
```

```python
import pandas as pd
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Загрузка данных
data = df_filtered.copy()
y = pd.DataFrame(data['salary_from_rub'])
y['salary_from_rub'] = y['salary_from_rub'].astype(int)

data.drop('salary_from_rub', axis=1, inplace=True)

data.dropna(axis=0, inplace=True)
data.drop(['index', 'v3_region_index', 'part', 'region'], axis=1, inplace=True)
numerical_columns = ['is_vahta', 'is_shift', 'is_distance', 'is_parttime', 'is_multiple', 'experience_id', 'is_junior', 'is_senior']
data[numerical_columns] = data[numerical_columns].astype(int)

# Преобразование столбца с регионом в категориальный признак
ct = ColumnTransformer([('encoder', OneHotEncoder(), [data.columns.get_loc('region_name')])], remainder='passthrough')
data_encoded = ct.fit_transform(data)
data_encoded_columns = ct.get_feature_names_out()

X = pd.DataFrame(data_encoded.toarray(), columns=data_encoded_columns)

# Удаление соответствующих строк из y
y = y.iloc[X.index]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Создание модели CatBoostRegressor и обучение
cat = CatBoostRegressor(
    learning_rate=0.001,
    depth=12,
    iterations=100000,
    random_seed=42,
    l2_leaf_reg=15,
    loss_function='MAE',
    eval_metric='MAPE',
    early_stopping_rounds=100
)
cat.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=100)

# Предсказание на обучающих данных
y_pred_train = cat.predict(X_train)

# Расчет MAPE на обучающих данных
mape_train = mean_absolute_percentage_error(y_train, y_pred_train)
print('MAPE (train):', mape_train)

# Предсказание на тестовых данных
y_pred_test = cat.predict(X_test)

# Расчет MAPE на тестовых данных
mape_test = mean_absolute_percentage_error(y_test, y_pred_test)
print('MAPE (test):', mape_test)
```

```python
# Вывод значимости признаков
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': cat.feature_importances_})
feature_importance.sort_values('Importance', ascending=False, inplace=True)
print(feature_importance)
```

```python
feature_importance_dict = {}

for index, row in feature_importance.iterrows():
    feature = row['Feature']
    importance = row['Importance']
    if 'encoder__region_name_' in feature:
        feature = feature.replace('encoder__region_name_', '')
        feature_importance_dict[feature] = importance

feature_importance_dict
```
